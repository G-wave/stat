# Sets Operations & Venn Diagram

```python
import matplotlib_venn as venn
import matplotlib.pyplot as plt
```

```python
print(set(range(5)))    # 
```

```python
print(set(range(2,6))) 
```

```python
print(set(range(2,16,3))) 
```

```python
A = set(range(2,7))
```

```python
B = {0, 2, 4, 6}
```

```python
venn.venn2([A,B], set_labels=('A','B'))
plt.show()
```

```python
C = set(range(5))
```

```python
venn.venn3([A,B,C], set_labels=('A','B','C'))
plt.show()
```
